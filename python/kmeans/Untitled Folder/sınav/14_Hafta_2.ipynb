```python
yillar=data["Year"].unique()
```

```python
yillar=sorted(yillar)
yillar
```

```python
ort_rating=[];
for i in yillar:
    x=data[ data["Year"]==i ]
    oran=sum(x["User Rating"])/len(x)
    ort_rating.append(oran)
```

```python
ort_rating
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.scatter(yillar,ort_rating, color="red" )
plt.plot(yillar,ort_rating, color="green")
plt.show()
```
